# Data Cleaning

```python
# Change the data type of the 'started_at' and 'ended_at' columns to datetime
data['started_at'] = pd.to_datetime(data['started_at'])
data['ended_at'] = pd.to_datetime(data['ended_at'])
```

```python
# Drop rows containing missing values
data = data.dropna()
```

```python
# Check data shape after drop missing values
data.shape
```

# Data Transforming

```python
# Added a 'day_name' column containing the name of the day
data['day_name'] = data['started_at'].dt.day_name()
```

```python
# Added a 'duration_minutes' column containing the duration in minutes
data['duration_minutes'] = (data['ended_at'] - data['started_at']).dt.total_seconds() / 60
data['duration_minutes'] = data['duration_minutes'].astype(int)
```

```python
data.sample(5)
```

```python
data.describe()
```

In the 'duration_minutes' column the minimum value is minus, so a cleaning process needs to be carried out for values that are less than zero or minus and for very high maximum values (more than 1440 minutes or 24 hours)

```python
negative_duration = (data['duration_minutes'] < 0).sum()
print('Number of duration below zero value :',negative_duration)
```

```python
duration_above_maximum_limit = (data['duration_minutes'] > 1440).sum()
print('Number of duration above 1440 minutes :',duration_above_maximum_limit)
```

```python
# Drop duration below zero and above 1440 minutes
data = data[(data['duration_minutes'] >= 0) & (data['duration_minutes'] <= 1440)]
```

```python
# Check min and max duration after drop duration < 0 minutes and >1440 minutes
print('Min Value : ', data['duration_minutes'].min())
print('Max Value : ', data['duration_minutes'].max())
```
